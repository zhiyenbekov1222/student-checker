```python
import pygame
from pygame.locals import *
import sys
import random
# инициализация пайгейм
pygame.init()
# шрифт и его размер 
font = pygame.font.SysFont("Arial", 40)


width, height = 500 , 500
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("RACEEER")

# изображений 
player_car = pygame.image.load("Player.png")
enemy_car = pygame.image.load("Enemy.png")
road = pygame.image.load("road.png")
coin_image = pygame.image.load("tenge.png")

# нужные для них расширений чтобы соответсовать ширине и длине
player_car = pygame.transform.scale(player_car, (50,100))
enemy_car = pygame.transform.scale(enemy_car, (50,100))
road = pygame.transform.scale(road, (width, height))
coin_image = pygame.transform.scale(coin_image, (30,30))

# музлон
pygame.mixer.music.load("back.wav")
pygame.mixer.music.play(-1, 0.0)

crash_sound = pygame.mixer.Sound("crash.wav")

# начальная координата игрока и врага
player_x = width // 2 - 25
player_y = height - 120
player_speed = 5

enemy_x = random.randint(50, width - 100)
enemy_y = -100
enemy_speed = 5

# цвета
red = (255,0,0)
black = (0,0,0)
green = (0, 255, 0)

score = 0
coin_count = 0
clock = pygame.time.Clock() # fps


# рисуем фигуру копейки и его расположений
coin_rect = pygame.Rect(random.randint(50, width-50), random.randint(-100, -30), 30, 30)

# после пройгрыша появится слова гейм овер 
def game_over():
    over_text = font.render("GAME OVER", True, red )
    screen.blit(over_text, (width//2 -80, height//2 - 20))
    pygame.display.update()



running = True
while running:
    # координаты
    screen.blit(road, (0,0))
    screen.blit(player_car, (player_x, player_y))
    screen.blit(enemy_car, (enemy_x, enemy_y))
    screen.blit(coin_image, coin_rect.topleft)
    
    # управление игрока
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and player_x > 0:
        player_x-=player_speed
    if keys[pygame.K_RIGHT] and player_x < width:
        player_x+=player_speed
    
    enemy_y+=enemy_speed
    
    # спец логика для того чтобы создать новый враг на рандомной У позиции а так же +1 балл и еще добавить скорость врага
    if enemy_y > height:
        enemy_y = -100
        enemy_x = random.randint(50, width - 100 )
        score +=1
        if score % 5 ==0:
            enemy_speed+=0.5
    
    coin_rect.y += enemy_speed
    
    # создание на рандомной позиции копейки
    if coin_rect.y > height:
        coin_rect.x = random.randint(50,width-50)
        coin_rect.y = random.randint(-100, -30)
    
    # если столкнулись с вражским машиной то игра заканчивается и возвращаем на функцию гейм овер
    player_rect = pygame.Rect(player_x, player_y, 50, 100)
    enemy_rect = pygame.Rect(enemy_x, enemy_y, 50 ,100)
    if player_rect.colliderect(enemy_rect):
        crash_sound.play()
        game_over()
        pygame.display.update()
        pygame.time.delay(2000)
        running = False
    
    # если попали на копейку то счет растет а так же копейка появляется на рандом позиции 
    if player_rect.colliderect(coin_rect):
        coin_count+=1
        coin_rect.x = random.randint(50, width - 50)
        coin_rect.y = random.randint(-100, -30)
    
    # текста
    score_text = font.render(f"Score: {score}", True, black)
    screen.blit(score_text, (10,10))
    
    coin_text = font.render(f"Coin: {coin_count}", True, red )
    screen.blit(coin_text, (width - 150, 10))

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        
    pygame.display.update()
    clock.tick(60)
pygame.quit()
sys.exit()
    
    
    
    
    
    

```
